```python
import numpy as np 
import pandas as pd 


def getindex71(a):
    final = a[1]*50+a[0]
    return  final

def getcord71(a):
    
    quo = a//50
    mod = np.mod(a,50)
    return [mod,quo]

def getindex72(a):
    final = 8640*a[5]+1440*a[4]+160*a[3]+32*a[2]+4*a[1]+a[0]
    print(final)
    return final

def getcord72(a):
    
    quo1 = a//8640
    mod1 = np.mod(a,8640)
    quo2 = mod1//1440
    mod2 = np.mod(mod1,1440)
    quo3 = mod2//160
    mod3 = np.mod(mod2,160)
    quo4 = mod3//32
    mod4 = np.mod(mod3,32)
    quo5 = mod4//4
    mod5 = np.mod(mod4,4)
    remain = a-quo1*8640-quo2*1440-quo3*160-quo4*32-quo5*4
    return [remain,quo5,quo4,quo3,quo2,quo1]
```
