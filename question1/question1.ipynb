```python
def coushu(x,k,target):
    # for any compeleted path in n*m matrix, at least n*(n+1)/2 is summed up for
    # the vertical direction in the matrix, so only need to find how may
    # "1" and "n"  is needed plus a remaining number to get target-n*(n+1)/2
    target-n*(n+1)/2  
    target = target - x*((x+1)/2)
    k = k-1
    need = math.floor((target - k)/x)
    
    if isinstance((target - k)/x,int):
        need = need -1
    need2 = k - need -1
    remain = int(target - need * x - need2)
    if remain> x:
        remain = target - (need+1)*x
        need = need +1
        need2 = need2-1 
    final  = [1]*need2  + [int(remain)] + [x]*need
    
    # output is a list contains needed "1" and "n" and a remained number
    return final
```
